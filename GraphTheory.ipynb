23. Construct a Heawood graph in a Circular layout. Check if the graph is Biaprtite or not. If it is Bipartite, print the two sets of vertices of the Bipartite graph.

```python
import matplotlib.pyplot as mp
from networkx.algorithms import bipartite
hea=nx.heawood_graph()
nx.draw_circular(hea,with_labels=True)
mp.show()
if nx.is_bipartite(hea)==True:
    print('Heawood Graph is bipartite!')
    set1,set2=bipartite.sets(hea)
    print('The first set is',set1,'\nThe second set is',set2)
else:
    print("Heawood Graph is not bipartite!")
```

A Clique is subgraph which is complete in nature.<br>The Clique number of a graph G is the highest of the orders of the possible cliques of a graph.

A Maximal Clique is a clique via which a particular set of vertices have reached their saturity.

An Independent set of vertices is a set of vertices which are not connected in the main graph.<br>A Maximal Independent set is a set in which no more vertices can be added without breaking the independent nature of the set.

```python
import networkx as nx
import matplotlib.pyplot as mp
G1=nx.wheel_graph(6)
nx.draw(G1,with_labels=True)
mp.show()
```

```python
#To get all the maximal cliques
list(nx.find_cliques(G1))
```

```python
#To get the maximum clique order which contains a particular node
nx.node_clique_number(G1,1)
```

```python
#To get the maximum clique order which contains a set of particular nodes
nx.node_clique_number(G1,[1,2])
```

```python
#To get clique number of a graph
nx.graph_clique_number(G1)
```

## Eulerian Graph

Eulerian Circuit
When there exists a path that traverses each edge exactly once such that the path begins and ends at the same vertex, the path is known as an Eulerian circuit and the graph is known as an Eulerian graph.
